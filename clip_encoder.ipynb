```python
from transformers import CLIPVisionModel
import torch
```

```python
device = "cpu"
```

```python
clip_model = CLIPVisionModel.from_pretrained("openai/clip-vit-large-patch14").to(device)
print(f"{sum(p.numel() for p in clip_model.parameters()):,d}")
```

```python
with torch.no_grad():
    clip_output = clip_model(torch.rand(1, 3, 224, 224))
clip_output
```
